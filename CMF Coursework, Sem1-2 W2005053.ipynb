```python
S0 = 288.67              # spot stock price
K =  360.80             # strike
T = 0.25                  # maturity 
r = 0.01                 # risk free rate 
sig = 0.2               # standard deviation (volatility)
N = 3                   # number of periods or number of time steps  
payoff = "put"          # payoff
```

```python
dT = float(T) / N                             # Delta t
u = np.exp(sig * np.sqrt(dT))                 # up factor
d = 1.0 / u                                   # down factor
```

```python
S = np.zeros((N + 1, N + 1))
S[0, 0] = S0
z = 1
for t in range(1, N + 1): #looping forwards, from 1 to N
    for i in range(z):  #looping forwards, from 0 to z-1
        S[i, t] = S[i, t-1] * u
        S[i+1, t] = S[i, t-1] * d
    z += 1  # same as z=z+1
```

```python
a = np.exp(r * dT)    # risk free compound return
p = (a - d)/ (u - d)  # risk neutral up probability
q = 1.0 - p           # risk neutral down probability
p
```

```python
S_T = S[:,-1]
V = np.zeros((N + 1, N + 1))
if payoff =="call":
    V[:,-1] = np.maximum(S_T-K, 0.0)
elif payoff =="put":
    V[:,-1] = np.maximum(K-S_T, 0.0)
V
```

```python
# for European Option
for j in range(N-1, -1, -1): # Column. looping backwards. From N-1 to 0
    for i in range(j+1):  # Row. looping forwards. From 0 to j
        V[i,j] = np.exp(-r*dT) * (p * V[i,j + 1] + q * V[i + 1,j + 1]) #the theoretical value at each node.
V
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T = 0.25    # Time to expiration (in years)
r = 0.01    # Risk-free interest rate
sigma = 0.2 # Volatility
S0 = 288.67 # Spot stock price
K = 360.80  # Strike price

# Number of simulations and time steps
num_simulations = 50  # You can adjust this based on your preference
num_time_steps = 1000

# Generate Monte Carlo paths
np.random.seed(42)
dt = T / num_time_steps
S_paths = np.zeros((num_simulations, num_time_steps + 1))
S_paths[:, 0] = S0

for i in range(1, num_time_steps + 1):
    Z = np.random.normal(0, 1, num_simulations)
    S_paths[:, i] = S_paths[:, i - 1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

# Plot Monte Carlo paths
plt.figure(figsize=(10, 6))
plt.title('Monte Carlo Simulation for Stock Price Paths')
plt.xlabel('Time Steps')
plt.ylabel('Stock Price')
for i in range(num_simulations):
    plt.plot(S_paths[i, :])

plt.axhline(y=K, color='r', linestyle='--', label='Strike Price')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

def euro_option_bs(S, K, T, r, vol, payoff):
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = d1 - vol * np.sqrt(T)

    if payoff == "call":
        option_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif payoff == "put":
        option_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    else:
        raise ValueError("Invalid option type. Use 'call' or 'put'.")

    return option_price

# Example usage
option_price_call = euro_option_bs(288.67, 360.80, 0.25, 0.01, 0.2, 'call')
option_price_put = euro_option_bs(288.67, 360.80, 0.25, 0.01, 0.2, 'put')

print(f"European Call Option Price: {option_price_call:.4f}")
print(f"European Put Option Price: {option_price_put:.4f}")
```

```python
def theta(S, K, T, r, vol, payoff):
    
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    N_d1_prime=1/np.sqrt(2 * np.pi) * np.exp(-d1**2/2)
    
    if payoff == "call":
        theta = - S * N_d1_prime * vol / (2 * np.sqrt(T)) - r * K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        theta = - S * N_d1_prime * vol / (2 * np.sqrt(T)) + r * K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return theta
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D

def rho(S, K, T, r, vol, payoff):
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    
    if payoff == "call":
        rho = T * K * np.exp(-r * T) * norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        rho = -T * K * np.exp(-r * T) * norm.cdf(-d2, 0.0, 1.0)
    
    return rho

# Given parameters
S_min = 220  # Minimum stock price
S_max = 420  # Maximum stock price
K = 360.80  # Strike price
r = 0.01  # Risk-free interest rate 
T = 3/12  # Time to expiration in years
sigma = 0.2  # Volatility

# Generate a range of stock prices
S = np.linspace(S_min, S_max, 51)

# Generate a range of time to expiry values
T_values = np.linspace(0.5, 2, 51)

# Initialize a matrix to store rho values
Rho = np.zeros((len(T_values), len(S)))

# Calculate rho values for different stock prices and time to expiry
for j in range(len(S)):
    for i in range(len(T_values)):
        Rho[i, j] = rho(S[j], K, T_values[i], r, sigma, 'call')

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
S, T_values = np.meshgrid(S, T_values)
surf = ax.plot_surface(S, T_values, Rho, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True)

# Set plot labels and title
ax.set_xlabel('Stock Price')
ax.set_ylabel('Time to Expiry')
ax.set_zlabel('Rho')
ax.set_title('Rho Surface')

# Add a colorbar for reference
fig.colorbar(surf, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
def vega(S, K, T, r, vol, payoff):
    
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    N_d1_prime=1/np.sqrt(2 * np.pi) * np.exp(-d1**2/2)
    vega = S * np.sqrt(T) * N_d1_prime
    
    return vega
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D

def vega(S, K, T, r, vol):
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    vega = S * norm.pdf(d1) * np.sqrt(T)
    return vega

# Given parameters
S_min = 220  # Minimum stock price
S_max = 420  # Maximum stock price
K = 360.80  # Strike price
r = 0.01  # Risk-free interest rate 
T = 3/12  # Time to expiration in years
sigma = 0.2  # Volatility

# Generate a range of stock prices
S = np.linspace(S_min, S_max, 51)

# Generate a range of time to expiry values
T_values = np.linspace(0.5, 2, 51)

# Initialize a matrix to store vega values
Vega = np.zeros((len(T_values), len(S)))

# Calculate vega values for different stock prices and time to expiry
for j in range(len(S)):
    for i in range(len(T_values)):
        Vega[i, j] = vega(S[j], K, T_values[i], r, sigma)

# Create a 3D surface plot for Vega
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
S, T_values = np.meshgrid(S, T_values)
surf = ax.plot_surface(S, T_values, Vega, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True)

# Set plot labels and title
ax.set_xlabel('Stock Price')
ax.set_ylabel('Time to Expiry')
ax.set_zlabel('Vega')
ax.set_title('Vega Surface')

# Add a colorbar for reference
fig.colorbar(surf, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as si
import yfinance as yf
```

```python
def euro_option_bs(S, K, T, r, vol, payoff):
    
    #S: spot price
    #K: strike price
    #T: time to maturity
    #r: risk free rate
    #vol: volatility of underlying asset
    #payoff: call or put
    
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        option_value = S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        option_value = - S * si.norm.cdf(-d1, 0.0, 1.0) + K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return option_value
```

```python
euro_option_bs(288.67, 360.80, 0.25, 0.01, 0.2, 'put')
```

```python
S = np.linspace(50,150,11)
cp = np.zeros((len(S),1))
for i in range(len(S)):
    cp[i] = euro_option_bs(S[i], 360.80, 0.25, 0.01, 0.2, 'put')
```
